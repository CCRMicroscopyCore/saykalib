```python
import os
import numpy as np
import pandas as pd
import re

from scipy import ndimage as nd 
from skimage.measure import label, regionprops
from tifffile import imread, imwrite
from tqdm import tqdm
```

```python
def roi_quant(lbl_img, embryo_mask):
    df = pd.DataFrame() 
    
    roi_id = [] 
    embryo_id = [] 
    
    roi_props = regionprops(lbl_img, intensity_image = embryo_mask)
    for roi in tqdm(range(len(roi_props))):
        roi_id.append(roi_props[roi].label)
        embryo_id.append(roi_props[roi].max_intensity)
        
    df['roi_id'] = roi_id
    df['embryo_id'] = embryo_id
    
    return df

def main(base_path, output_path):
    set_list = ['04_26', '04_27', '04_28'] 
    for set_name in set_list:
        set_path = os.path.join(base_path, set_name) 
        output_path2 = os.path.join(output_path, set_name)
        
        tmp_list = os.listdir(set_path) 
        img_list = [] 
        for tmp in tmp_list: 
            if re.search('_cp_masks.tif', tmp):
                img_list.append(tmp) 
        
        print(img_list)

        for img_name in img_list: 
            print(img_name.replace('_ruby_cp_masks.tif', '')) 
            img_path = os.path.join(set_path, img_name) 
            img = imread(img_path) 
            
            embryo_path = os.path.join(set_path, img_name.replace('ruby_cp_masks.tif', 'embryo.tif'))
            embryo_img = imread(embryo_path)
            
            img_out_path = os.path.join(output_path2, img_name.replace('_ruby_cp_masks.tif', ''))
            
            for z_plane in tqdm(range(img.shape[0])):
                img_frame = img[z_plane, :, :] 
                embryo_frame = embryo_img[z_plane, :, :]
                
                df = roi_quant(img_frame, embryo_frame) 
                df_name = img_name.replace('_ruby_cp_masks.tif', '_embryo_') + str(z_plane) + '.csv' 
                tmp_path = os.path.join(img_out_path, df_name) 
                df.to_csv(tmp_path, header = True)
```

```python
base_path = '/Volumes/LECIMAGE/Analysis/[NCI] [LGI] Sergio Ruiz/Bechara Saykali/images/dhb_embryo_04'
output_path = '/Volumes/LECIMAGE/Analysis/[NCI] [LGI] Sergio Ruiz/Bechara Saykali/output/dhb_embryo_04'
```

```python
main(base_path, output_path) 
```
